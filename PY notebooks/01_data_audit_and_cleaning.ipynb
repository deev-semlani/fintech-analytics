```python
#1: Lib import
import pandas as pd
import numpy as np
import os

print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
```

```python
PATH = r"../data/raw data/"

df_users = pd.read_csv(PATH + "users_copy.csv")
df_loans = pd.read_csv(PATH + "loans_copy.csv")
df_repayments = pd.read_csv(PATH + "repayments_copy.csv")
df_collections = pd.read_csv(PATH + "collections_copy.csv")


print("All files loaded")
print(f"Users: {len(df_users)} rows")
print(f"Loans: {len(df_loans)} rows")
print(f"Repayments: {len(df_repayments)} rows")
print(f"Collections: {len(df_collections)} rows")
```

```python
#cell 3a: Having a look at data

print("USERS")
print(df_users.head())

print("\nLOANS")
print(df_loans.head())

print("\nREPAYMENTS")
print(df_repayments.head())

print("\nCOLLECTIONS")
print(df_collections.head())
```

```python
#cell 3b: understanding data
df_users.info()
df_loans.info()
df_repayments.info()
df_collections.info()
```

```python
#4: Clean users table (Fix age column)

print(df_users[df_users['age'].apply(lambda x: not str(x).isdigit())])
```

```python
#5: Fix age column- mapping text to numbers

age_map = {
    'forty two'   : 42,
    'thirty one'  : 31,
    'twenty eight': 28,
    'twenty three': 23
}

df_users['age'] = df_users['age'].replace(age_map)
df_users['age'] = pd.to_numeric(df_users['age'], errors='coerce')


print(df_users['age'].dtype)
print(df_users['age'].isna().sum()) #just verifying
```

```python
#6 Fixing city name

city_map = {'Bombay': 'Mumbai','Bangalore' : 'Bengaluru', 'NEW DELHI': 'Delhi', 'bangalore': 'Bengaluru', 'mumbai': 'Mumbai', 'new delhi' : 'Delhi', 'delhi': 'Delhi'}

df_users['city']= df_users['city'].replace(city_map)

print(df_users['city'].value_counts())
```

```python
# Cell 7: Fix employment type

print(df_users['employment_type'].value_counts())
print()
job = {'slef-employed':'Self-Employed', 'Self Employeed':'Self-Employed','Freelncer':'Freelancer', 'Gig worker': 'Gig Worker', 'salaried':'Salaried'}
df_users['employment_type']= df_users['employment_type'].replace(job)
print (df_users['employment_type'].value_counts())
```

```python
# Cell 8: Fix date formats

df_users['signup_date'] = pd.to_datetime(df_users['signup_date'], dayfirst=True)

print(df_users['signup_date'].dtype)
print(df_users['signup_date'].head())
```

```python
# Cell 9: missing credit scores

print(df_users['credit_score'].isna().sum())
print(df_users.groupby('employment_type')['credit_score'].apply(lambda x: x.isna().sum()))
```

```python
# Cell 9 continued: missing credit scores

df_users['credit_score'] = df_users.groupby('employment_type')['credit_score'].transform(lambda x: x.fillna(x.median()))

print(df_users['credit_score'].isna().sum())
print(df_users.groupby('employment_type')['credit_score'].median())
```

```python
# Cell 10: Saving cleaned users

PROCESSED_PATH = r"../data/processed/"

df_users.to_csv(PROCESSED_PATH + "cleaned_users.csv", index=False)
```

```python
# Cell 11: Audit loans table

print(df_loans.shape)
print(df_loans.dtypes)
print(df_loans.isnull().sum())
```

```python
# Cell 12: correct date format

df_loans['disbursement_date'] = pd.to_datetime(df_loans['disbursement_date'], dayfirst=True)
print(df_loans.dtypes)
```

```python
#cell 13: solve the -ve procesing fee

df_loans.loc[df_loans['processing_fee']< 0,[ 'processing_fee'] ]= df_loans.loc[df_loans['processing_fee'] < 0, ['processing_fee']] * -1
```

```python
#cell 14: droping a row to solve small loan aount issue

df_loans['loan_amount'] = pd.to_numeric(df_loans['loan_amount'], errors='coerce')
df_loans = df_loans[df_loans['loan_amount']>1000]
print(df_loans.shape)


```

```python
#cell 15: casing issue in loan_pirpose

print (df_loans['loan_purpose'].value_counts())

print()

purpose = {'medical emergency':'Medical Emergency','WEDDING':'Wedding','education':'Education','HOME REPAIR':'Home Repair','travel':'Travel'}
df_loans ['loan_purpose']= df_loans ['loan_purpose'].replace(purpose)
print (df_loans['loan_purpose'].value_counts())
```

```python
#cell 16: Saving this csv

df_loans.to_csv (PROCESSED_PATH + 'cleaned_loans.csv', index=False)
```

```python
#cell 17: repayments data cleaning

print(df_repayments.shape)
print(df_repayments.dtypes)
print ()
print(df_repayments.isnull().sum())
```

```python
#cell: 18 fill empty values

df_repayments['payment_date'] = df_repayments['payment_date'].fillna('31-12-2099')
df_repayments['payment_mode'] = df_repayments['payment_mode'].fillna('Not Paid')
df_repayments['days_late']    = df_repayments['days_late'].fillna(181)
```

```python
#cell 19: change datatypes


df_repayments['payment_date']= pd.to_datetime (df_repayments['payment_date'], dayfirst=True)
df_repayments['due_date'] = pd.to_datetime (df_repayments ['due_date'], format= 'mixed' , dayfirst=True)
df_repayments.dtypes
```

```python
#cell 20: duplicate and casing problem solved

df_repayments.duplicated().sum()
df_repayments = df_repayments.drop_duplicates
df_repayments.shape


df_repayments ['payment_mode'] = df_repayments ['payment_mode'].str.title()
df_repayments ['payment_mode'] = df_repayments ['payment_mode'].str.strip()

map= {'Auto Debit':'Auto-Debit'}
df_repayments ['payment_mode']= df_repayments ['payment_mode'].replace(map)
df_repayments['payment_mode'].value_counts()
```

```python
#cell 21:saveit!
df_repayments.to_csv(PROCESSED_PATH + 'cleaned_repayments.csv' , index= False)
```

```python
#cell 22: new data csv info 
print (df_collections.shape)
print (df_collections.dtypes)
print (df_collections.isnull().sum())
```

```python
#cell 23: formatting

df_collections ['contact_date']=  pd.to_datetime (df_collections ['contact_date'], dayfirst=True)
df_collections ['promise_to_pay_date']=  pd.to_datetime (df_collections ['promise_to_pay_date'], dayfirst=True)
print (df_collections.dtypes)
```

```python
#cell32 checking for any casing issues in respose and contact_mode
print (df_collections ['contact_mode'].value_counts())
print ()
print (df_collections ['response'].value_counts())
```

```python
#cell33: save it

df_collections.to_csv(PROCESSED_PATH + 'cleaned_collections.csv', index= False)
```
