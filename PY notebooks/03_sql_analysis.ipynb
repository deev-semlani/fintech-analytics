```python
#only req once to create sqlite db when setting up project fresh
import sqlite3
import pandas as pd


PROCESSED_PATH = r"../data/processed/"

df_users = pd.read_csv(PROCESSED_PATH + "cleaned_users.csv")
df_loans = pd.read_csv(PROCESSED_PATH + "cleaned_loans.csv")
df_repayments = pd.read_csv(PROCESSED_PATH + "cleaned_repayments.csv")
df_collections = pd.read_csv(PROCESSED_PATH + "cleaned_collections.csv")


conn = sqlite3.connect(r"../sql files/fatakpay.db")

#putting all df as table in sqlite
df_users.to_sql('users', conn, if_exists='replace', index=False)
df_loans.to_sql('loans', conn, if_exists='replace', index=False)
df_repayments.to_sql('repayments', conn, if_exists='replace', index=False)
df_collections.to_sql('collections', conn, if_exists='replace', index=False)

print("Database created at sql/fatakpay.db")
print(f"Users: {len(df_users)} rows loaded")
print(f"Loans: {len(df_loans)} rows loaded")
print(f"Repayments: {len(df_repayments)} rows loaded")
print(f"Collections: {len(df_collections)} rows loaded")
```
